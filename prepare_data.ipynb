```python
import json
import pandas as pd
```

```python
ellipsis = pd.read_csv('data/raw/ellipsis corpus.csv')
ellipsis.head()
```

```python
# add id column with ellipsis_{index}
ellipsis['id'] = 'ellipsis_' + ellipsis.index.astype(str)
ellipsis.dropna()
ellipsis.head()
```

```python
ellipsis.to_csv('data/ellipsis.csv', index=False)
```

```python
with open('data/raw/disrpt.json', encoding="utf-8") as f:
    disrpt = json.load(f)
# rename keys from 0 and 1 to "disrpt_{index}"
for i in range(len(disrpt)):
    disrpt[str(i)]['id'] = 'disrpt_' + str(i)
disrpt
```

```python
with open('data/disrpt_with_id.json', 'w', encoding="utf-8") as f:
    json.dump(disrpt, f, ensure_ascii=False, indent=4)
```
